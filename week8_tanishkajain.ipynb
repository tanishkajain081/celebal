````python

# Calculator

def calculator(expression: str) -> str:
    """
    Evaluate a mathematical expression.

    Parameters:
        expression (str): Mathematical expression (e.g., "5+10*2")

    Returns:
        str: Result of the calculation or an error message.
    """
    try:
        result = eval(expression)
        return str(result)
    except ZeroDivisionError:
        return "Error: Division by zero."
    except SyntaxError:
        return "Error: Invalid mathematical expression."
    except Exception as e:
        return f"Error: {e}"
````

````python

#   Keyword Extractor


import string

def extract_keywords(text: str) -> list:
    """
    Extract keywords from the given text.

    Parameters:
        text (str): Input sentence.

    Returns:
        list: List of unique keywords.
    """
    try:
        # Convert to lowercase
        text = text.lower()

        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Split into words
        words = text.split()

        # Basic stop words
        stop_words = {
            "the", "is", "are", "and", "for", "with", "this", "that",
            "from", "have", "has", "will", "into", "your", "their",
            "about", "been", "were", "was", "than", "then", "they",
            "them", "also", "very", "over", "under", "what", "when",
            "where", "which", "while", "whose", "would", "could",
            "should", "after", "before", "there", "here", "these",
            "those", "you", "your", "ours", "our"
        }

        # Extract unique keywords
        keywords = []
        for word in words:
            if len(word) > 4 and word not in stop_words:
                if word not in keywords:
                    keywords.append(word)

        return keywords[:5]

    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
def agent(query: str):
    query_lower = query.lower()

    try:
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)
            return {
                "type": "calculation",
                "result": result
            }

        elif "keywords" in query_lower:
            text = (query_lower
                    .replace("extract","")
                    .replace("keywords","")
                    .replace("from","")
                    .strip())
            result = extract_keywords(text)
            return {
                "type": "keywords",
                "result": result
            }

        else:
            return {
                "type": "general",
                "result": f"General response: {query}"
            }

    except Exception as e:
        return {
            "type": "error",
            "result": str(e)
        }
````

## 📦 Expected Output Format

  "result": ...
}

````python
# ==========================================
# 🎯 INTERACTIVE MODE
# ==========================================

print("=" * 50)
print("🤖 AI Agent Started")
print("Type your query below.")
print("Type 'exit' to stop the program.")
print("=" * 50)

while True:
    user_input = input("\nEnter query: ").strip()

    if user_input.lower() == "exit":
        print("\n👋 Thank you for using the AI Agent!")
        break

    try:
        response = agent(user_input)

        print("\nResponse:")
        print(response)

    except Exception as e:
        print(f"\nError: {e}")
````
